```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelBinarizer
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
from sklearn.metrics import mean_squared_error
from keras import backend as K

def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true))) 

test_LGBT = pd.read_csv("LGBT_test.csv",header=None)
test_XGBT = pd.read_csv("XGBoost_test.csv",header=None)
test_CABT = pd.read_csv("CatBoost_test.csv",header=None)


y = pd.read_csv("y_test.csv",header=None)

ensemble = (test_LGBT + test_XGBT + test_CABT)/3
LGBT = (test_LGBT*1.0) 
XGBT = (test_XGBT*1.0) 
CABT = (test_CABT*1.0) 


print("LGBT_RMSE:",np.sqrt(mean_squared_error(y, LGBT)))
print("XGBT_RMSE:",np.sqrt(mean_squared_error(y, XGBT)))
print("CABT_RMSE:",np.sqrt(mean_squared_error(y, CABT)))
print()
print("RMSE:",np.sqrt(mean_squared_error(y, ensemble)))
print()
print("RMSE:",19.156933616910674)
```
